```python
import pandas as pd

train_file = "/content/Datasets/Heart_Disease/Training_set_heart.csv"
test_file = "/content/Datasets/Heart_Disease/Testing_set_heart.csv"

train_df = pd.read_csv(train_file)
test_df = pd.read_csv(test_file)

print("Training Set:")
display(train_df.head())

print("Testing Set:")
display(test_df.head())

#
```

```python
print(train_df.info())

print(train_df.isnull().sum())

print(train_df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x="target", data=train_df)
plt.title("Heart Disease: Presence vs Absence (Training Set)")
plt.show()

plt.figure(figsize=(12, 8))
sns.heatmap(train_df.corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap (Training Set)")
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

X_train = train_df.drop("target", axis=1)
y_train = train_df["target"]

model= LogisticRegression(max_iter=1000) # I changed to 1000 to prevent an error
model.fit(X_train, y_train)
```

```python
X_test = test_df
y_pred = model.predict(X_test)

test_df["predicted_target"]=y_pred
display(test_df.head())



```

```python
sns.countplot(x="predicted_target", data=test_df)
plt.title("Predicted Heart Disease Counts (Testing Set)")
plt.show()
```
